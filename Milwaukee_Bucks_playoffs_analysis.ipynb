# **Milwaukee Bucks playoffs analysis**

### This interactive python notebook consists of an analysis through the data about the Milwaukee Bucks NBA team playoff runs of the: 19-20, 20-21, 21-22 and 22-23 seasons. More specifically about the team's First Round performance.

### Why First Round? Because in the 19-20 season Milwaukee beat Orlando by 4 x 0, in the 20-21 season Milwaukee beat Miami by 4 x 1, in the 21-22 season Milwaukee beat Chicago by 4 x 1. **But** in the 22-23 season Milwaukee lost to Miami, 1 x 4.
